```python
top_brands = {}
brands = ['volkswagen', 'bmw', 'mercedes_benz', 'audi', 'opel', 'ford' ]
```

```python
for b in brands:
    selected_rows = autos[autos['brand'] == b]
    price_mean = selected_rows['price'].mean()
    top_brands[b] = price_mean
print(top_brands)
```

#The most popular selling cars on this eBay site are in two pricing tiers:
#High-price tier (Audi, Mercedes-Benz, BMW)
#Mid-price tier (Volkswagen, Ford, Opel)

```python
mileage_km = {}
for b in brands:
    selected_row = autos[autos['brand'] == b]
    mileage_mean = selected_row['odometer_km'].mean()
    mileage_km[b] = mileage_mean
print(mileage_km)
```

```python
#series constructor
bmp_series = pd.Series(top_brands)
print(bmp_series)
```

```python
df = pd.DataFrame(bmp_series, columns=['mean_price'])
df
```

```python
bmp_2series = pd.Series(mileage_km)
df['mean_mileage'] = bmp_2series
df
```
