```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0],
recent_grads.head(),
recent_grads.tail())
recent_grads.describe()
```

```python
recent_grads = recent_grads.dropna()
```

# Scatter plots

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full_time and Median')
ax = recent_grads.plot(x='Part_time', y='Median', kind='scatter')
#Is there any link between the number of full-time employees and median salary?
#There is a slight trend that students that work full time in general earn more money(25000-41000), part- timers(20000-40000)
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median')
#Do students that majored in subjects that were majority female make more money?
#the correlation between the students that majored in subjects were the majority were women and their salaries exist.
#We can notice a thend which implies: the more women as share of total graduated with that major the less money students make
```

```python
ax = recent_grads.plot(x='Total', y='Median', kind='scatter', title='Total and Median')
#Do students in more popular majors make more money?
#The points are mainly concentrated in the bottom left corner (instead of top right)
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen and Unemployment_rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men and Median')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women and Median')
```

Rank - Rank by median earnings (the dataset is ordered by this column).
Major_code - Major code.
Major - Major description.
Major_category - Category of major.
Total - Total number of people with major.
Sample_size - Sample size (unweighted) of full-time.
Men - Male graduates.
Women - Female graduates.
ShareWomen - Women as share of total.
Employed - Number employed.
Median - Median salary of full-time, year-round workers.
Low_wage_jobs - Number in low-wage service jobs.
Full_time - Number employed 35 hours or more.
Part_time - Number employed less than 35 hours.

# Histograms

```python
recent_grads['Sample_size'].plot(kind='hist')
```

```python
ax = recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

```python
recent_grads['Median'].hist(bins=11, range=(0,120000), color='pink')
plt.xlabel('Median salary')
plt.title('Distribution of salary')
#In general most recent graduates earn around 40000
```

```python
recent_grads['Employed'].hist(bins=7, range=(0,310000), color='purple')
plt.xlabel('Number of employed')
plt.ylabel('frequency')
plt.title('Number of employed students')

```

```python
cols = ['Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
fig = plt.figure(figsize=(7,20))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    ax = recent_grads[cols[i]].plot(kind='hist', rot=30)
```
